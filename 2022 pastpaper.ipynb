```python
import numpy as np
import matplotlib.pyplot as plt

# Evaluate truncated series with N terms
numbers = [10,100,1000] # a list of N to evaluate
numbers_iter = iter(numbers) # set a variable to equal the iterations of the list "numbers"
approx = [1] # create a list to store approximations in for estimated error
index = 0 # create an index variable for iterating though approximations for estimated error
for i in numbers:
   # Creating an array of 
   a = next(numbers_iter) 
   N = np.linspace(1, a, a)
   # print(N)
   List = []
   for i in N:
      List.append(1/(i**2))
      # print(List)
   ans = np.sqrt(6*sum(List))
   approx.append(ans)
   print(f'The approxiamtion with {a} terms is {ans}')
   
   #Errors
   #Estimated error
   ea = (ans - approx[index])/ans
   index += 1
   print(f'The estimated error is {ea}')
   
   #True error
   pi = np.pi
   true_error = pi - ans
   print(f'The true error is {true_error}\n')
   
```
