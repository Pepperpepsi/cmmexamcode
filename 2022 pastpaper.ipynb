```python
import numpy as np

# Evaluate truncated series with N terms
numbers = [10,100,1000] # a list of N to evaluate
approx = [1] # create a list to store approximations in for estimated error
index = 0 # create an index variable for iterating though approximations for estimated error
for i in numbers:
   N = np.linspace(1, i, i)
   List = []
   for i in N:
      List.append(1/(i**2))
      # print(List)
   ans = np.sqrt(6*sum(List))
   approx.append(ans)
   print(f'The approxiamtion is {ans}')
   
   #Errors
   #Estimated error
   ea = (ans - approx[index])/ans
   index += 1
   print(f'The estimated error is {ea}')
   
   #True error
   pi = np.pi
   true_error = pi - ans
   print(f'The true error is {true_error}\n')
   
```

The stability of a numerical method is how well it conforms to the exact solution and whether it diverges at large intervals (high values of x) given a variety of conditions. A numericla method can be described as stable if it does not diverge and conform to the exact solution over a large range of different functions and step sizes. 

The forward Euler method is explict. It takes the current state of the system and computes the next state of the system using a function. The backwards Euler method is implicit, where an equation involving both the current and later state of a system needs to be solved.

```python
import numpy as np

iterations = np.arange(1,100000,1)
f = lambda xn: (1/np.sin(xn))+(1/4)
f_min = lambda x: f(x) - x
xl = 1
xu = 4

def bracketing_seq(f,xl,xu):
    while True:
        xr = (xl+xu)/2
        yield xr
        a = f(xl)*f(xr)
        # print(a)
        if a < 0:
            xu = xr
        elif a > 0:
            xl = xr
        else:
            print(f'Exact solution found: {xr}')
midpoints = bracketing_seq(f_min, xl, xu)

for i in iterations:
    print(next(midpoints))



#graphing
# x = next(midpoints)
# x_coords = np.linspace(0,4,100)
# x_coords = np.delete(x_coords,0,0)
# y_coords = []
# for i in x_coords:
#     y = f(i)
#     y_coords.append(y)
# x_coords = x_coords.tolist()
# x_coords.insert(0,0)
# y_coords.insert(0,0)
# x_coords2 = y_coords
# y_coords2 = y_coords
# plt.plot(x_coords, y_coords)
# plt.plot(x_coords2, y_coords2)
# plt.scatter(x,x, color="red")
# plt.xlim(0,4)
# plt.ylim(-10,10)
# plt.show()
```
